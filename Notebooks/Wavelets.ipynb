The Haar wavelet can be used to great effect when analysing time series. Using the definition of the Haar, we can constructe the wavelet directl from step functions.

```python
haar = Step() - Step(0.5,weight=2)

ninf = Step()*3

fig, ax = plt.subplots(nrows=2)
x = np.arange(0,1,0.01)
haar.plot(ax=ax[0])
#ax[0].step(x,haar(x))
prod = haar*ninf
ax[0].step(x,(prod>>0.25)(x))

ninf.plot(ax=ax[1])
#ax[1].step(x,ninf(x))
```

```python
def generate_step_func(samples, x_vals):
    hfuncs = np.zeros(len(x_vals))

    for i in range(samples):
        signs = random.choice([-1,1])
        shifts = random.randint(0,5)
        hfuncs += signs*np.heaviside(x_vals-shifts,1)
        
    return hfuncs

h = generate_step_func()
```
